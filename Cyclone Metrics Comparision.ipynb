```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('peak_int.csv',parse_dates=['TIME'])
df2 = pd.read_csv('int_rate.csv',parse_dates=['TIME'])
df3 = pd.read_csv('int_duration.csv',parse_dates=['TIME'])
```

```python
dataset = pd.merge(pd.merge(df1,df2,on=['YEAR','TIME']),df3,on=['YEAR','TIME'])
```

```python
dataset['Prediction'] = dataset['AveInt']*dataset['Duration']*4 + 33
```

```python
ax = dataset.plot('TIME','Prediction')
dataset.plot('TIME','VMAX',ax=ax)
plt.title('r=0.7149')
plt.show()
```

```python
np.corrcoef(dataset['Prediction'], dataset['VMAX'])
```
