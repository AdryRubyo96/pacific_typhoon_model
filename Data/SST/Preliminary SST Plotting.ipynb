```python
# LIBRARY IMPORT
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Unpack the dataset
int_region = pd.read_csv('int_region_sst.csv')
global_sst = pd.read_csv('global_sst.csv')
```

```python
dataset = int_region.merge(global_sst,on='year')
```

```python
dataset.columns = (['year','int_region_sst','global_sst'])
```

```python
dataset['delta'] = dataset['int_region_sst'] - dataset['global_sst']
```

```python
dataset.plot('year','delta')
plt.show()
```
